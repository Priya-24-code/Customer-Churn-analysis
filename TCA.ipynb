```python
def conv(value):
     if value == 1:
         return "yes"
     else:
         return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Crosstab (count)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert to percentage
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(5,4))

# Step 4: Add percentage labels inside bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Step 5: Titles and labels
plt.title("Churn by SeniorCitizen (%)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
n_cols = 3 
n_rows = (len(cols)  + n_cols -1) // n_cols  # calculate number of rows needed

# create grid (3 rows, 3 cols)
fig, axes = plt.subplots(n_rows,n_cols, figsize=(15, n_rows * 4))

# flatten axes for easy looping
axes = axes.flatten()

# loop through columns
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue='Churn')
    
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

    # add labels on bars
    for container in axes[i].containers:
        axes[i].bar_label(container)

# adjust layout
plt.tight_layout()
plt.show()
```
